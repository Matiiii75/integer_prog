```python
import os 
import subprocess

dossier_instances = os.listdir("../instances/")
print(len(dossier_instances))
```

```python
# permet de lancer automatiquement toutes les instances pour différents types de gen col : j'ai juste a modifier le parametre 0,1,2

alphas = [0.25, 0.5, 0.75, 0.9]

for i in range(3): #pr chaque para
    for instance in dossier_instances: 
        if instance == "uniform_26.inst": continue
        print("traitement de : ", instance, "pour para ", i, " (0:pricing mip, 1: sac, 2: sac+stab)")
        
        if i == 2 : #cas où on fait de la stabilization 
            for a in alphas: 
                print("alpha = ", a)
                cmd = ["./projet_ip",  f"../instances/{instance}", str(i), "1"]
                subprocess.run(cmd, input = f"{a}\n", text=True)
        else : 
            cmd = ["./projet_ip",  f"../instances/{instance}", str(i), "1"]
            subprocess.run(cmd)
        print()

```

```python
# je l'ai codé pr pouvoir lancer la stabilisation avec différents parametre facilement

for instance in dossier_instances:
    if(instance == "uniform_26.inst"): continue
    print("traitement de :", instance)
    cmd = ["./projet_ip", f"../instances/{instance}", "2", "1"]
    subprocess.run(cmd)
    print()
```
