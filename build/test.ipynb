```python
import os 
import subprocess

dossier_instances = sorted(os.listdir("../instances/"))
print(dossier_instances)
```

```python
# permet de lancer automatiquement toutes les instances pour diff√©rents types de gen col : j'ai juste a modifier le parametre 0,1,2
from datetime import datetime

alphas = [0.25, 0.5, 0.75, 0.9]
banned_instances = ["uniform_26.inst", "uniform_27.inst", "uniform_31.inst"]

for i in range(3): #pr chaque para
    for instance in dossier_instances: 
        temps_courant = datetime.now()
        if instance in banned_instances : continue
        print("traitement de : ", instance, "pour para ", i, " (0:pricing mip, 1: sac, 2: sac+stab)")
        print("Heure : ", temps_courant)
        if i == 2 : #cas o√π on fait de la stabilization 
            for a in alphas: 
                print("alpha = ", a)
                cmd = ["./projet_ip",  f"../instances/{instance}", str(i), "1"]
                subprocess.run(cmd, input = f"{a}\n", text=True)
        else : 
            cmd = ["./projet_ip",  f"../instances/{instance}", str(i), "1"]
            subprocess.run(cmd)
        print()

```

```python
# je l'ai cod√© pr pouvoir lancer la stabilisation avec diff√©rents parametre facilement

for instance in dossier_instances:
    if(instance == "uniform_26.inst"): continue
    print("traitement de :", instance)
    cmd = ["./projet_ip", f"../instances/{instance}", "2", "0"]
    subprocess.run(cmd, input=f"0.9\n", text=True)
    print()
```
